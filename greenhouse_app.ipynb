```python
# app.py

import streamlit as st
import supabase as sp
import pandas as pd
from supabase import create_client, Client
```

```python
# --- 1. Verbindung zu Supabase herstellen ---
# Lädt die Zugangsdaten sicher aus der secrets.toml Datei
@st.cache_resource
def init_connection():
    url = st.secrets["SUPABASE_URL"]
    key = st.secrets["SUPABASE_KEY"]
    return create_client(url, key)

supabase = init_connection()
```

```python
# --- 2. Funktion zum Abfragen der Daten ---
# Diese Funktion führt eine SQL-Abfrage aus und gibt das Ergebnis als DataFrame zurück
# st.cache_data sorgt dafür, dass die Daten zwischengespeichert werden und nicht bei jeder Interaktion neu geladen werden müssen
@st.cache_data
def run_query(query):
    response = supabase.rpc('execute_sql', {'sql_query': query}).execute()
    data = response.data
    return pd.DataFrame(data)
```
